## Importemos Datos de entrada

```python
dataframe = pd.read_csv(r"comprar_alquilar.csv")
dataframe.head(10)
# REFERENCIAS:
# ingresos y gastos son mensuales de 1 personas o 2 si están casados.
# trabajo: 0-sin trabajo 1-autonomo 2-asalariado 3-empresario 4-Autonomos 5-Asalariados 6-Autonomo y Asalariado 7-Empresario y Autonomo 8 Empresarios o empresario y autónomo
# estado_civil: 0-soltero 1-casado 2-divorciado
# hijos: Cantidad de hijos menores (no trabajan)
# comprar: 0-mejor alquilar 1-Comprar casa
# hipoteca fija a 30 años con interes
```

# Preprocesamos los datos

Creamos 2 columnas nuevas. En una agrupamos los gastos mensuales. En la otra el monto a financiar para comprar la casa

```python
dataframe['gastos'] = (dataframe['gastos_comunes'] + dataframe['gastos_otros'] + dataframe['pago_coche'])
dataframe['financiar'] = dataframe['vivienda'] - dataframe['ahorros']
dataframe.drop(['gastos_comunes', 'gastos_otros', 'pago_coche'], axis=1).head(10)
```

```python
reduced = dataframe.drop(['gastos_comunes', 'gastos_otros', 'pago_coche'], axis=1)
reduced.describe()
```

### Ejemplo: Graficar 2 de las Características

```python
used_features2 = [
    "ingresos",
    "financiar"
]
X = dataframe[used_features2].values
y = dataframe["comprar"]

fig, ax = plt.subplots()

ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='RdBu')
ax.set_title('Naive Bayes Model', size=14)

xlim = (2000, 9000)
ylim = (100000, 700000)

xg = np.linspace(xlim[0], xlim[1], 40)
yg = np.linspace(ylim[0], ylim[1], 30)
xx, yy = np.meshgrid(xg, yg)
Xgrid = np.vstack([xx.ravel(), yy.ravel()]).T

for label, color in enumerate(['red', 'blue']):
    mask = (y == label)
    mu, std = X[mask].mean(0), X[mask].std(0)
    P = np.exp(-0.5 * (Xgrid - mu) ** 2 / std ** 2).prod(1)
    Pm = np.ma.masked_array(P, P < 0.03)
    ax.pcolorfast(xg, yg, Pm.reshape(xx.shape), alpha=0.5,
                  cmap=color.title() + 's')
    ax.contour(xx, yy, P.reshape(xx.shape),
               levels=[0.01, 0.1, 0.5, 0.9],
               colors=color, alpha=0.2)

ax.set(xlim=xlim, ylim=ylim)
```

Vemos que cuantos más ingresos tiene la familia y menor es la cantidad a financiar, mejora la opción de Comprar (en azul). De otra manera, convendrá alquilar (rojo)

# Creamos el modelo de Gaussian Naive Bayes

Dividimos en un set de Entrenamiento y otro de Test con el 20% de las entradas

```python
# Split dataset in training and test datasets
X_train, X_test = train_test_split(dataframe, test_size=0.2, random_state=6)
y_train = X_train["comprar"]
y_test = X_test["comprar"]
```

Entrenamos el modelo

```python
# Instantiate the classifier
gnb = GaussianNB()

# Train classifier
gnb.fit(X_train[used_features].values, y_train)

y_pred = gnb.predict(X_test[used_features])
```

# Resultados

```python
print('Precisión en el set de Entrenamiento: {:.2f}'
      .format(gnb.score(X_train[used_features], y_train)))
print('Precisión en el set de Test: {:.2f}'
      .format(gnb.score(X_test[used_features], y_test)))
```

```python
# Print Test results
print("Total de Muestras en Test: {}\nFallos: {}"
.format(
    X_test.shape[0],
    (y_test != y_pred).sum()
))
```

Las métricas de evaluación son herramientas utilizadas para medir y comprender el rendimiento de un modelo o algoritmo en tareas específicas, como la clasificación o la regresión. Estas métricas proporcionan una forma fácil de interpretar y comparar cómo está funcionando un modelo en relación con los resultados reales o las expectativas.

En términos menos técnicos, las métricas de evaluación son como "medidores" que nos ayudan a entender cuán bien o cuán mal está haciendo nuestro modelo en diferentes aspectos. Piensa en ellas como formas de responder preguntas clave sobre el rendimiento de tu modelo, Existen diferentes formas de evaluar los resultados de un algoritmo:

> **Metricas numéricas:** Precisión, Exhaustividad, Exactitud, F1...
> **Reprensentaciones Gráficas:** Curva ROC, Curva PR...

Todas las métricas mencionadas y existentes se fundamentan en un concepto crucial conocido como "matriz de confusión", que ofrece una visión esclarecedora:

> **Verdadero Positivo:** Es un ejemplo que se ha clasificado correctente como perteneciente a la clase "1" o positivo.
> **Verdadero Negativo:** Es un ejemplo que se ha clasificado correctamente como perteneciente a la clase "0" o negativo.
> **Falso Positivo:** Es un ejemplo que se ha clasificado como perteneciente a la clase "1" pero realmente pertenece a la clase "0".
> **Falso Negatico:** Es un ejemplo que se ha clasificado como perteneciente a la clase "0" pero realmente pertenece a la clase "1".

```python
# Creamos la matris de confusión
cm = confusion_matrix(y_test, y_pred)
print(cm)
# Visualizar la matriz de confusión
disp = ConfusionMatrixDisplay(cm, display_labels=gnb.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

"Al evaluar mis predicciones en relación con las etiquetas reales, observamos los siguientes resultados:

**Verdaderos Positivos (21 casos):** En situaciones en las que los valores o flujos estaban etiquetados como 'es mejor alquilar' en nuestra base de datos, mi algoritmo logró predecir correctamente que estos también eran adecuados para alquilar. Este acierto se traduce en 21 casos en los que las predicciones coincidieron con las preferencias reales.

**Falsos Positivos (0 casos):** Sin embargo, para los flujos de datos o valores que ya estaban etiquetados como 'alquilar' en nuestro subconjunto de datos, mi algoritmo hizo predicciones erróneas. Específicamente, en ningún caso, mi algoritmo sugirió que deberían comprarse en lugar de alquilarse. Estos son considerados falsos positivos, ya que las predicciones contradicen las etiquetas reales.

**Verdaderos Negativos (13 casos):** En relación con aquellos flujos o valores que demostraron ser más adecuados para la opción de compra en nuestro subconjunto, mi algoritmo también acertó. Se predijeron correctamente 13 casos en los que las preferencias para comprar coincidieron con las recomendaciones del algoritmo.

**Falsos Negativos (4 casos):** Por último, en algunas situaciones en las que los datos indicaban que era preferible comprar, mi algoritmo realizó predicciones equivocadas al sugerir que deberían alquilarse en lugar de comprarse. Estos 4 casos representan áreas de oportunidad para mejorar la precisión de las recomendaciones.

# Métricas.

```python
# Desglose de los valores en la matriz de confusión
tp = cm[0, 0]
tn = cm[1, 1]
fp = cm[0, 1]
fn = cm[1, 0]

print("Verdaderos Positivos:", tp)
print("Verdaderos Negativos:", tn)
print("Falsos Positivos:", fp)
print("Falsos Negativos:", fn)
```

* **Precisión:** Se calcula dividiendo el número de verdadero positivo por la suma de número de verdadero positivo y el número de falso posotivo.


```python
# Utilizamos la funsión precision_score() pasandole 2 parametros nuestras etiquetas y las prediciones de nuestro algoritmo.
print("presición: ", precision_score(y_test, y_pred))

# Otro forma de sacar la precisión
precision = tp / (tp + fp)
print(precision)
```

* **Exhaustividad (recall):** Se calcula dividiendo el número de verdadero positivos por la suma del número de verdadero positivo y el número de falso negativo.

```python
# utilizamos la funsión recall_score() pasandole 2 parametros nuestras etiquetas y las prediciones de nuestro algoritmo.
print("Exhaustividad: ", recall_score(y_test, y_pred))

# Otra forma de sacar la Exhaustividad
Exhaustividad = tp / (tp + fn)
print(Exhaustividad)
```

* **Exactitud (accuracy):** Suma del número de verdadero positivos y el número de verdaderos negativos dividido por el número total de ejemplos

```python
# utilizamos la funsión accuracy_score() pasandole 2 parametros nuestras etiquetas y las prediciones de nuestro algoritmo.
print('Exactitud: ', accuracy_score(y_test, y_pred))

# Total de muestras.
total_samples = tp + tn + fp + fn
accuracy = (tp + tn) / total_samples
print(accuracy)
```

* **F1-score:** Se puede interpretar como un promedio ponderado de la precisión y la exactitud (recall), donde el f1-score alcanza su mejor valor en 1 y el peor puntaje en 0.
* La contribución relativa de la precisión y la exactitud al F1-score es la misma [1].

```python
# utilizamos la funsión f1_score() pasandole 2 parametros nuestras etiquetas y las prediciones de nuestro algoritmo.
print('F1 score: ', f1_score(y_test, y_pred))

f1_score = 2 * (precision * Exhaustividad) / (precision + Exhaustividad)
print(f1_score)
```

* **Especificidad:** La especificidad es una métrica que mide la capacidad de un modelo para identificar correctamente los casos negativos. En otras palabras, cuán bueno es el modelo en predecir que una instancia pertenece a la clase negativa cuando realmente lo es. Una alta especificidad indica que el modelo tiene una alta tasa de verdaderos negativos (TN) y baja tasa de falsos positivos (FP), lo que significa que el modelo es capaz de discriminar eficazmente entre los casos negativos. Por otro lado, una baja especificidad indica que el modelo está clasificando incorrectamente muchos casos negativos como positivos, lo que puede ser problemático en aplicaciones donde los casos negativos son críticos.

```python
especificidad = tn / (tn + fp)
print("Especificidad:", especificidad)
```

Para analizar cómo nuestro algoritmo está funcionando de manera visual, vamos a explorar la Curva ROC (Receiver Operating Characteristic), una herramienta que nos proporciona información sobre su desempeño.

 * **Curva Roc (Receiver Operating Cahracteristic):** Suele representar la tasa de verdaderos positivos en el eje Y y la una tasa de falso positivos en ek eje X. Esto significa que la **esquina superiro izquierdo de la gráfica es el punto "ideal"**, una tasa de falso positivo de cero y una tasa de verdadero positivos de uno. Como consecuencia un area grande debajo de la curva (AUC) suele ser mejor [1].
